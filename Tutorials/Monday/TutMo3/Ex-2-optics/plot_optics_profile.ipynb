```python
%matplotlib inline
import numpy as np
import pylab as plt
import sys
import os

#Plot contents of twiss_profiles.txt which is produced by get_twiss_profiles() in PyZgoubi
FL=os.getcwd()
s, beta_y, alpha_y, disp_y, beta_z, alpha_z = np.loadtxt('%s/%s'%(FL,'twiss_profiles0.txt'),skiprows=2,
                                                 usecols=(0,3,4,6,9,10),unpack=True)


plt.plot(s,beta_y, 'k',label=r'$\beta_y$')
plt.plot(s,beta_z, 'r',label=r'$\beta_z$')
plt.xlabel('s [cm]')
plt.ylabel(r'$\beta$')
plt.title('Optics in KEK 150 MeV FFA cell')
plt.legend()
plt.show()

plt.plot(s,alpha_y, 'k',label=r'$\alpha_y$')
plt.plot(s,alpha_z, 'r',label=r'$\alpha_z$')
plt.xlabel('s [cm]')
plt.ylabel(r'$\alpha$')
plt.legend()
plt.show()

plt.plot(s,disp_y, 'k')
plt.xlabel('s [cm]')
plt.ylabel(r'$\eta$')
plt.legend()
plt.show()
```
