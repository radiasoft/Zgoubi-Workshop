```python
%matplotlib inline
from zgoubi import core
import numpy as np
import pylab as plt
import sys
import os

#make use of PyZgoubi Results class
#r = core.Results()
#obtain coordinates from the zgoubi.fai file. The Results class figures out which column of data to extract.
#data = r.get_track('zgoubi_singlepass.plt', ['Y','T', 'S', 'Bz'])
#data = np.transpose(data)
#y, t, s, bz = data

#Alternatively, use numpy loadtxt to read the columns directly.
#y,t,z,p are the radial coordinate and angle and the vertical coordinate and angle, respectively.
FL=os.getcwd()
y, t, s, bz = np.loadtxt('%s/%s'%(FL,'zgoubi_singlepass.plt'),skiprows=2,usecols=(9,10,13,24),unpack=True)

plt.plot(s,y, 'k-')
plt.xlabel('s [cm]')
plt.ylabel('r [cm]')
plt.title('Closed orbit')
plt.show()

plt.plot(s,t, 'k-')
plt.xlabel('s [cm]')
plt.ylabel('r\' [mrad]')
plt.title('Closed orbit (angle)')
plt.show()


plt.plot(s,bz, 'k-')
plt.xlabel('s [cm]')
plt.ylabel(r'$B_z$ [kG]')
plt.title('Field along closed orbit')
plt.show()
```
