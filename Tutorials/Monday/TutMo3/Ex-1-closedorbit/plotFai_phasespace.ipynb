```python
%matplotlib inline
from zgoubi import core
import numpy as np
import pylab as plt
import sys
import os

#make use of PyZgoubi Results class
#r = core.Results()
#obtain coordinates from the zgoubi.fai file. The Results class figures out which column of data to extract.
#coords = r.get_track('zgoubi_iter1.fai', ['Y','T', 'Z', 'P'])
#coords = np.transpose(coords)

#Alternatively, use numpy loadtxt to read the columns directly.
FL=os.getcwd()
coords = np.loadtxt('%s/%s'%(FL,'zgoubi_iter1.fai'),skiprows=2,usecols=(9,10,11,12),unpack=True)

#y,t,z,p are the horizontal coordinate and angle and the vertical coordinate and angle, respectively.
y, t, z, p = coords

y_mean = np.mean(y)
t_mean = np.mean(t)

print "mean x,x' (cm)",y_mean, t_mean
print "max - min x, x' (cm) ",max(y)-min(y)
plt.plot(y,t, 'ko')
plt.plot([y_mean], [t_mean],'ro',label='{<x>,<x\'>}')
plt.xlabel('x [cm]')
plt.ylabel('x\' [mrad]')
plt.title('Horizontal phase space')
plt.legend()
plt.show()
```
