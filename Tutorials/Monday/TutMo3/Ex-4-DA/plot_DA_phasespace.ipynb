```python
%matplotlib inline
import numpy as np
import pylab as plt
import sys
import os

#Plot contents of twiss_profiles.txt which is produced by get_twiss_profiles() in PyZgoubi
FL=os.getcwd()
indices_all, y, t = np.loadtxt('%s/%s'%(FL,'dynamic_aperture_coords.txt'),skiprows=0,
                                                 usecols=(0,1,2),unpack=True)
#number of amplitudes in file
namp = int(indices_all[-1])

for i in range(namp):
    indices = np.where(indices_all == i)
    plt.plot(y[indices],t[indices],'o')
    
plt.xlabel('Y [cm]')
plt.ylabel('T [mrad]')
plt.title('Phase spaces at various amplitudes')
plt.show()

```
