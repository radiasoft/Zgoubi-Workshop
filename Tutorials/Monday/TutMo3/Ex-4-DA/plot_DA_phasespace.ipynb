```python
%matplotlib inline
import numpy as np
import pylab as plt
import sys
import os

#Plot contents of twiss_profiles.txt which is produced by get_twiss_profiles() in PyZgoubi
FL=os.getcwd()
indices_all, y, t = np.loadtxt('%s/%s'%(FL,'dynamic_aperture_coords.txt'),skiprows=0,
                                                 usecols=(0,1,2),unpack=True)

amp, tune_h, tune_v = np.loadtxt('%s/%s'%(FL,'tune_amplitude.txt'),skiprows=1,
                                                 usecols=(0,1,2),unpack=True)
#number of amplitudes in file
namp = int(indices_all[-1])

for i in range(namp):
    indices = np.where(indices_all == i)
    plt.scatter(y[indices],t[indices],marker='.')
    
plt.xlabel('Y [cm]')
plt.ylabel('T [mrad]')
plt.title('Phase spaces at various amplitudes')
plt.show()


plt.plot(1e6*amp, tune_h,'ko-')
plt.xlabel('single particle emit. [pi mm mrad]')
plt.ylabel('cell tune')
plt.show()

```
